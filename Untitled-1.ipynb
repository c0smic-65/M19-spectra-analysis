```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
K = 0.307075  # MeV mol^-1 cm^2
z = 1  # Charge of muon
Z_A = 0.5357  # Z/A for scintillator
me = 0.51099892e-3  # Electron mass (GeV/c^2)
c = 3e8  # Speed of light (m/s)
I = 63e-9  # Mean excitation energy (GeV)
rho = 1.032  # Density of scintillator (g/cm^3)

# Muon energies to plot (GeV)
E_mu = np.logspace(-1, 3, 100)

# Calculate beta, gamma
beta = np.sqrt(1 - (me**2) / (E_mu**2))
gamma = 1 / np.sqrt(1 - beta**2)

# Maximum transferable kinetic energy
T_max = 2 * me * beta**2 * gamma**2 / (1 + 2 * gamma * me / 0.105658)

# Bethe-Bloch formula
dEdx = (
    K
    * z**2
    * Z_A
    / beta**2
    * (0.5 * np.log(2 * me * beta**2 * gamma**2 * T_max / I**2) - beta**2)
)

# Plot dE/dx vs muon energy
plt.semilogx(E_mu, dEdx * 1e3, "-r", lw=2)
plt.xlabel("Muon Energy (GeV)")
plt.ylabel("dE/dx (MeV cm$^2$/g)")
plt.title("Energy Loss of Muons in Scintillator")
plt.grid(True, which="both")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
K = 0.307075  # MeV cm^2/g
Z = 6
A = 14
rho = 1.032  # g/cm^3
I = 0.0745  # MeV
m_e = 0.511  # MeV/c^2
m_mu = 105.7  # MeV/c^2
x = 2  # cm


def bethe_bloch(beta):
    if beta >= 1.0:
        beta = 0.9999  # avoid division by zero for gamma
    gamma = 1 / np.sqrt(1 - beta**2)
    Tmax = (
        2 * m_e * beta**2 * gamma**2 / (1 + 2 * gamma * m_e / m_mu + (m_e / m_mu) ** 2)
    )
    dEdx = (
        K
        * (Z / A)
        * (1 / beta**2)
        * (0.5 * np.log(2 * m_e * beta**2 * gamma**2 * Tmax / I**2) - beta**2)
    )
    return dEdx * rho  # Convert to MeV/cm


# Generate beta values from 0.85 to just below 1
beta_values = np.linspace(0.85, 0.9999, 150)

# Calculate delta E for each beta value
delta_E_values = [bethe_bloch(beta) * x for beta in beta_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(
    beta_values, delta_E_values, label="Energy Loss in 2 cm of Scintillator", color="b"
)
plt.xlabel("Beta Factor (β)")
plt.ylabel("Energy Loss ΔE (MeV)")
plt.title("Energy Loss of Muons in Scintillator as a Function of β")
plt.legend()
plt.grid(True)
plt.show()
```
